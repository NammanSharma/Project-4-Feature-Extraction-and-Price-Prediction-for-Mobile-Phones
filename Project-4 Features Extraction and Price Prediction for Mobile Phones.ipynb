## **Handling the Outliers.**

```python
for i in num_cols:
    Q1=df[i].quantile(0.25)
    print("The value of first quantile for column",i,"is",Q1)
    Q2=df[i].quantile(0.50)
    print("The value of second quantile for column",i,"is",Q2)
    Q3=df[i].quantile(0.75)
    print("The value of third quantile for column",i,"is",Q3)
    IQR=Q3-Q1
    print("The Interquartile Range is",IQR)
    LL=Q1-1.5*(Q3-Q1)
    print("The Lower Limit for column",i,"is",LL)
    UL=Q3+1.5*(Q3-Q1)
    print("The Upper Limit for column",i,"is",UL)
    n=df[(df[i]<LL)|(df[i]>UL)].count()[0]
    print("The number of outliers for column",i,"is",n)
    df[i]=np.where((df[i]<LL)|(df[i]>UL),df[i].mean(),df[i])
    n=df[(df[i]<LL)|(df[i]>UL)].count()[0]
    print("The number of outliers after imputation is",n)
    print("")
    print("")
```

# Performing the Machine Learning Algoritms to determine the best Model for Mobile Price Prediction.

## Model Building.

```python
df
```

```python
categorical_columns=df.select_dtypes(include=["object"]).columns
```

```python
categorical_columns
```

## Here we are converting data into numeric form as machine only underatands numbers.


### Using LabelEncoder.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for i in categorical_columns:
    df[i]=LabelEncoder().fit_transform(df[i])
```

```python
df  # Here all the columns are converted into numeric form.
```

## Split data into x and y (80% training and 20% testing).

```python
x=df.drop("Price",axis=1)
```

```python
x
```

```python
y=df["Price"]
```

```python
y
```

## Seperate data into train and test.

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=8)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

## Applying different Models.

## Model 1- Linear Regression.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
model1=LinearRegression()
model1.fit(x_train,y_train)
y_pred=model1.predict(x_test)
```

```python
# Evaluating model performance.

score=model1.score(x_test,y_test).round(3)
print(f'Model Score: {score}')

# Calculate and print the Mean Squared Error
mse = mean_squared_error(y_test, y_pred).round(3)
print('Mean Squared Error:', mse)

# Calculate and print the R-squared score
r2 = r2_score(y_test, y_pred).round(3)
print('R-squared:', r2)

# Calculate and print the mean absolute error
mae = mean_absolute_error(y_test, y_pred).round(3)
print('Mean Absolute Error:', mae)
```

## Model 2- Decision Tree.

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model2=DecisionTreeRegressor(random_state=42)
model2.fit(x_train,y_train)
y_pred=model2.predict(x_test)
```

```python
# Evaluating model performance.

score=model2.score(x_test,y_test).round(3)
print(f'Model Score: {score}')

# Calculate and print the Mean Squared Error
mse = mean_squared_error(y_test, y_pred).round(3)
print('Mean Squared Error:', mse)

# Calculate and print the R-squared score
r2 = r2_score(y_test, y_pred).round(3)
print('R-squared:', r2)

# Calculate and print the mean absolute error
mae = mean_absolute_error(y_test, y_pred).round(3)
print('Mean Absolute Error:', mae)
```

## Model 3- Random Forest Regressor.

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model3=RandomForestRegressor()
model3.fit(x_train,y_train)
y_pred=model3.predict(x_test)
```

```python
# Evaluating model performance.

score=model3.score(x_test,y_test).round(3)
print(f'Model Score: {score}')

# Calculate and print the Mean Squared Error
mse = mean_squared_error(y_test, y_pred).round(3)
print('Mean Squared Error:', mse)

# Calculate and print the R-squared score
r2 = r2_score(y_test, y_pred).round(3)
print('R-squared:', r2)

# Calculate and print the mean absolute error
mae = mean_absolute_error(y_test, y_pred).round(3)
print('Mean Absolute Error:', mae)
```

## Model 4 - Gradient Boosting.

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
model4=GradientBoostingRegressor()
model4.fit(x_train,y_train)
y_pred=model4.predict(x_test)
```

```python
# Evaluating model performance.

score=model4.score(x_test,y_test).round(3)
print(f'Model Score: {score}')

# Calculate and print the Mean Squared Error
mse = mean_squared_error(y_test, y_pred).round(3)
print('Mean Squared Error:', mse)

# Calculate and print the R-squared score
r2 = r2_score(y_test, y_pred).round(3)
print('R-squared:', r2)

# Calculate and print the mean absolute error
mae = mean_absolute_error(y_test, y_pred).round(3)
print('Mean Absolute Error:', mae)
```

## Comparing the MAE, MSE and R-Squared of the models for final selection of model:

```python
models = ['Linear Regression', 'Decision Tree','Random Forest Regressor','Gradient Boosting']
mae_values = [2251.133, 1243.727, 1185.841, 1374.117]
r2_values = [ 0.627, 0.73, 0.822, 0.813]
mse_values=[9626596.343, 6968800.869, 4599793.187, 4831614.807]
model_comparison = pd.DataFrame({'Model': models,'MAE': mae_values,'R-Squared': r2_values,'MSE':mse_values})
print(model_comparison)
best_model = model_comparison.loc[model_comparison['MAE'].idxmin()] #idxmin() finds the index of the minimum MAE value.
print("  ")
print("  ")
print("Best Model based on MAE, R-Squared and MSE:")
print(best_model)
```
